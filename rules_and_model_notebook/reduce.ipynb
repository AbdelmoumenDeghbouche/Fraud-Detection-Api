```python
import pandas as pd

df=pd.read_csv('rules.csv')
new=pd.DataFrame()
# Extract information from the 'Antecedent' column into new columns
new['Gender'] = df['antecedents'].str.extract(r'gender_(\w+)')
new['Age'] = df['antecedents'].str.extract(r'Age Category_(\w+)')
new['Distance'] = df['antecedents'].str.extract(r'Distance Category_(\w+)')
new['Transaction'] = df['antecedents'].str.extract(r'Transaction Category_(\w+)')
new['Population'] = df['antecedents'].str.extract(r'Population Category_(\w+)')
new['Amount'] = df['antecedents'].str.extract(r'Amt Category_(\w+)')
new['Category'] = df['antecedents'].str.extract(r'category_(\w+)')


new.to_csv('rules_simple.csv', index=False)
```
